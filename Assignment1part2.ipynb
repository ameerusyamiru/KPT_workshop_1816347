```python
img = cv2.imread(path+"coins.jpg")
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
greyimage = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
imgBlur = cv2.GaussianBlur(greyimage,(5,5),0)

plt.figure(figsize=[20,20])
plt.subplot(131);plt.imshow(img)
plt.subplot(132);plt.imshow(greyimage)
plt.subplot(133);plt.imshow(imgBlur)
```

```python
imgCanny1 = cv2.Canny(imgBlur,30,600)
plt.imshow(imgCanny1)
# imgCanny1 = cv2.Canny(greyimage,30,90)
# plt.imshow(imgCanny1)

contours, hierarchy = cv2.findContours(imgCanny1, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(img, contours, -1, (0, 255, 255), 10)
# plt.imshow(img)
```
